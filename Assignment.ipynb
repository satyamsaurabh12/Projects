```python
# What are iterators , iterable & generators in Python ?
# iterators :->
#             An object that helps you go through a collection (like a list or tuple) item by item.
#             After all items are accessed, calling next() again will raise a StopIteration error, indicating there are no more items.    
fruits = ['apple', 'banana', 'cherry']
iterator = iter(fruits)

print(next(iterator))  
print(next(iterator)) 
print(next(iterator))

# Iterable :-> An object you can loop through to get its items, like a list, tuple, or string.
numbers = [1, 2, 3]
for number in numbers:
    print(number)
# Generators :-> A function that yields values one by one and can be used in a loop to access items sequentially.
def count_up_to_three():
    yield "Satyam"
    yield [1,2,3,45]
    yield (1,3,7,9)

counter = count_up_to_three()

print(next(counter)) 
print(next(counter))
print(next(counter))  
```

```python
# What is encapsulation? Explain it with an example.

 # It refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, or class, and restricting access to some of the object's components. 
# Encapsulation allows for controlled access to the data through methods, ensuring that the data is safe
class Bank:
     def __init__(self,Name,account_number,balance):
         self.holder=Name
         self.__account_number = account_number
         self.__balance = balance
     def deposit(self,ammount):
         self.__balance = self.__balance+ammount
     def printing_accounnumber(self):
         print(f"The holder {self.holder} has account_Number is :{self.__account_number}")
     def withdraw(self,ammount):
         self.__balance=self.__balance - ammount
     def printing_balance(self):
         print(f"The {self.holder} has ammount is :{self.__balance}")

cus1=Bank("Satyam","122cs0056",10000)
cus1.printing_accounnumber()
print("\nThe ammount before the deposit is :\n")
cus1.printing_balance()
cus1.deposit(1000)
print("\nThe ammount after the deposit is :\n")
cus1.printing_balance()
```

```python
# 4.Create a simple banking system using object-oriented concepts in Python. Design classes representing 
# different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, 
# and balance inquiry. Utilize inheritance to manage different account types efficiently.


# Tasks:
# Define base class(es) for bank accounts with common attributes and methods
class BankAccount:
    def __init__(self, AccountHolder, AccountNumber, Balance):
        self.AccountHolder = AccountHolder
        self._AccountNumber = AccountNumber
        self._Balance = Balance
    
    def deposit_amount(self, amount):
        self._Balance += amount
        print(f"Deposited amount: {amount}. New balance is: {self._Balance}")
    
    def withdraw_amount(self, amount):
        if self._Balance < amount:
            print("Your account has insufficient balance for the withdrawal.")
        else:
            self._Balance -= amount
            print(f"Successfully withdrew {amount}. New balance is: {self._Balance}")
    
    def balance_inquiry(self):
        print(f"Account Holder: {self.AccountHolder}, Account Number: {self._AccountNumber}, Balance: {self._Balance}")

class SavingAccount(BankAccount):
    def __init__(self, AccountHolder, AccountNumber, Balance, interest_rate):
        super().__init__(AccountHolder, AccountNumber, Balance)
        self.interest_rate = interest_rate
    
    def with_interest_rate(self):
        interest = (self._Balance * self.interest_rate) / 100
        self._Balance += interest
        print(f"Interest applied. New balance is: {self._Balance}")

class CheckingAccount(BankAccount):
    def __init__(self, AccountHolder, AccountNumber, Balance, checking_amount):
        super().__init__(AccountHolder, AccountNumber, Balance)
        self.checking_amount = checking_amount
    
    def apply_checking_amount(self):
        self._Balance -= self.checking_amount
        print(f"Checked amount: {self.checking_amount}. New balance is: {self._Balance}")


# Testing the classes
Cus1 = BankAccount("Satyam", "122cs0056", 1000)
Cus1.deposit_amount(2000)
Cus1.withdraw_amount(1500)
Cus1.balance_inquiry()

Cus2 = SavingAccount("Saurabh", "122cs0091", 2000, 5)
Cus2.with_interest_rate()
Cus2.balance_inquiry()

Cus3 = CheckingAccount("Shashank", "122cs0090", 3000, 30)
Cus3.apply_checking_amount()
Cus3.balance_inquiry()
           
        
# Implement subclasses for specific account types (e.g. SavingAccount,CheckingAccount)
# Provide methods for deposit,withdraw ,and balances inquiry in each subclass
# Test the banking system by creating instances of different account types and performing transaction
```

```python
# 6.Write a code for Restaurant Management System Using OOPS:
# Create a MenuItem class that has attributes such as name, description, price, and category.
# Implement methods to add a new menu item, update, menu item information, and remove a menu item 
# from the menu.
# Use Encapsulation to hide the menu item's unique identification number.
# Inherit from the MenuItem class to create a FoodItem class and a BeverageItem class, each with their own 
# specific attributes and methods
class MenuItem:
    def __init__(self, name, description, price, category):
        self.name = name
        self.description = description
        self.price = price
        self.category = category
        self.__unique_item = hash(name)  # Private attribute
        self.mydisc = {self.__unique_item: [self.name, self.description, self.price, self.category]}

    def add_new_item(self, name, description, price, category):
        unique_id = hash(name)
        self.mydisc[unique_id] = [name, description, price, category]

    def remove_item(self, name):
        unique_id = hash(name)
        if unique_id in self.mydisc:
            del self.mydisc[unique_id]
            print("This item is removed from Menu Item.")
            return
        print("This item is not available in this menu.")

    def print_disc(self):
        print(self.mydisc)

class FoodItem(MenuItem):
    def __init__(self,name,description,price,category,foodtime):
        super().__init__(name,description,price,category)
        self.foodtime = foodtime
    def printing_foodtime(self):
        print(f"{self.name} has food timing {self.foodtime} and its description {self.description} and its price {self.price} and its category {self.category}")

class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, beverage_size):
        super().__init__(name, description, price, category)
        self.beverage_size = beverage_size

    def print_beverage_details(self):
        print(f"Beverage Item: {self.name}, Size: {self.beverage_size}, Description: {self.description}, Price: {self.price}, Category: {self.category}")

    
    
MenIt = MenuItem("rice", "best quality", 345, "Basmati")
MenIt.add_new_item("Bread", "best Quality", 456, "Wheat")

MenIt.add_new_item("Brea1", "best Quality", 456, "Wheat")

MenIt.print_disc()
MenIt.remove_item("Brea1")

MenIt.print_disc()

print("\n For fooditem is.......\n")
Food = FoodItem("Panner","best Quality","547","cow","Dinner")
Food.printing_foodtime()
beverage = BeverageItem("Butter","best quality","543","cow","500mg")
beverage.print_beverage_details()
```

```python
# 7.Write a code for  Hotel Management System using OOPS :
# Create a Room class that has attributes such as room number, room type, rate, and availability (private).
# Implement methods to book a room, check in a guest, and check out a guest.
# Use Encapsulation to hide the room's unique identification number.
# Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own 
# specific attributes and methods
class Room:
    def __init__(self, room_number, room_type, rate, availability):
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = availability

    def check_availability(self):
        if self.__availability == 1:
            print(f"Room {self.room_number} is available.")
        else:
            print(f"Sorry! Room {self.room_number} is already occupied.")

    def book_room(self):
        if self.__availability == 1:
            print(f"Room {self.room_number} booked successfully.")
            self.__availability = 0
        else:
            print(f"Sorry! Room {self.room_number} is already occupied.")

    def check_in_guest(self, name):
        if self.__availability == 1:
            print(f"Guest '{name}' checked into Room {self.room_number}.")
            self.__availability = 0 
        else:
            print(f"Sorry! Room {self.room_number} is already occupied.")

    def check_out_guest(self, name):
        print(f"Guest '{name}' checked out from Room {self.room_number}.")
        self.__availability = 1 

class SuiteRoom(Room):
    def __init__(self, room_number, rate, availability, suite_type):
        super().__init__(room_number, "Suite", rate, availability)
        self.suite_type = suite_type 

    def print_suite_details(self):
        self.check_availability()
        print(f"Suite Type: {self.suite_type}, Room Number: {self.room_number}, Rate: {self.rate}")

class StandardRoom(Room):
    def __init__(self, room_number, rate, availability, bed_type):
        super().__init__(room_number, "Standard", rate, availability)
        self.bed_type = bed_type  

    def print_standard_details(self):
        self.check_availability()
        print(f"Bed Type: {self.bed_type}, Room Number: {self.room_number}, Rate: {self.rate}")


room101 = Room(101, "Deluxe", 200, 1)
room101.check_availability()
room101.book_room()
room101.check_availability()
room101.check_out_guest("John Doe")
room101.check_availability()

# SuiteRoom management
suite1001 = SuiteRoom(1001, 500, 1, "Presidential Suite")
suite1001.print_suite_details()
suite1001.book_room()

# StandardRoom management
standard102 = StandardRoom(102, 150, 1, "King Bed")
standard102.print_standard_details()
standard102.check_in_guest("Alice Smith")
```

```python
# 8.Write a code for  Fitness Club Management System using OOPS:

# + Create a Member class that has attributes such as name, age, membership type, and membership status (private).
# + Implement methods to register a new member, renew a membership, and cancel a membership.
# + Use Encapsulation to hide the member's unique identification number.
# + Inherit from the Member class to create a FamilyMember class and an IndividualMember class, Each with 
# their own specific attributes and methods
class Member:
    def __init__(self, name, age, membership_type, membership_status):
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = membership_status  
        self.__unique_id = hash(name)  

    def register_member(self):
        if self.__membership_status == "Inactive":
            self.__membership_status = "Active"
            print(f"{self.name} has been successfully registered.")
        else:
            print(f"{self.name} is already an active member.")

    def renew_membership(self):
        if self.__membership_status == "Active":
            print(f"{self.name}'s membership has been renewed.")
        else:
            print(f"{self.name} is not an active member. Please register first.")

    def cancel_membership(self):
        if self.__membership_status == "Active":
            self.__membership_status = "Inactive"
            print(f"{self.name}'s membership has been canceled.")
        else:
            print(f"{self.name} is not an active member.")

    def check_membership_status(self):
        print(f"Membership Status for {self.name}: {self.__membership_status}")

    def get_unique_id(self):
        return self.__unique_id 

class FamilyMember(Member):
    def __init__(self, name, age, membership_type, membership_status, family_size):
        super().__init__(name, age, membership_type, membership_status)
        self.family_size = family_size 

    def print_family_details(self):
        self.check_membership_status()
        print(f"Family Member: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Family Size: {self.family_size}")

class IndividualMember(Member):
    def __init__(self, name, age, membership_type, membership_status, personal_trainer=False):
        super().__init__(name, age, membership_type, membership_status)
        self.personal_trainer = personal_trainer  

    def assign_trainer(self):
        if self.personal_trainer:
            print(f"{self.name} already has a personal trainer assigned.")
        else:
            self.personal_trainer = True
            print(f"A personal trainer has been assigned to {self.name}.")

    def print_individual_details(self):
        self.check_membership_status()
        trainer_status = "Yes" if self.personal_trainer else "No"
        print(f"Individual Member: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Personal Trainer: {trainer_status}")


member1 = Member("John Doe", 30, "Gold", "Inactive")
member1.register_member()
member1.renew_membership()
member1.cancel_membership()
member1.check_membership_status()


family_member = FamilyMember("Jane Doe", 35, "Platinum", "Active", 4)
family_member.print_family_details()


individual_member = IndividualMember("Alice Smith", 28, "Silver", "Inactive")
individual_member.register_member()
individual_member.assign_trainer()
individual_member.print_individual_details()
```

```python
# 9.Write a code for  Event Management System using OOPS
# Create an Event class that has attributes such as name, date, time, location, and list of attendees (private).
# Implement methods to create a new event, add or remove attendees, and get the total number of attendees.
# Use Encapsulation to hide the event's unique  identification number.
# Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, Each with their own 
# specific attributes and methods
class Event:
    def __init__(self, name, date, time, location):
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []  
        self.__event_id = hash(name + date + time + location) 

    def create_event(self):
        print(f"Event '{self.name}' created successfully for {self.date} at {self.time} in {self.location}.")

    def add_attendee(self, attendee_name):
        self.__attendees.append(attendee_name)
        print(f"'{attendee_name}' has been added to the event '{self.name}'.")

    def remove_attendee(self, attendee_name):
        if attendee_name in self.__attendees:
            self.__attendees.remove(attendee_name)
            print(f"'{attendee_name}' has been removed from the event '{self.name}'.")
        else:
            print(f"'{attendee_name}' is not in the attendee list for the event '{self.name}'.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def get_attendee_list(self):
        return self.__attendees

    def get_event_id(self):
        return self.__event_id  

class PrivateEvent(Event):
    def __init__(self, name, date, time, location, invitation_required=True):
        super().__init__(name, date, time, location)
        self.invitation_required = invitation_required 

    def check_invitation_status(self):
        if self.invitation_required:
            print(f"Invitation is required for the private event '{self.name}'.")
        else:
            print(f"No invitation is required for the private event '{self.name}'.")

class PublicEvent(Event):
    def __init__(self, name, date, time, location, max_capacity):
        super().__init__(name, date, time, location)
        self.max_capacity = max_capacity  

    def add_attendee(self, attendee_name):
        if self.get_total_attendees() < self.max_capacity:
            super().add_attendee(attendee_name)
        else:
            print(f"Cannot add '{attendee_name}' to the event '{self.name}'. Maximum capacity of {self.max_capacity} reached.")

    def check_capacity(self):
        current_attendees = self.get_total_attendees()
        print(f"The event '{self.name}' currently has {current_attendees}/{self.max_capacity} attendees.")


event1 = Event("Annual Conference", "2024-09-10", "10:00 AM", "Convention Center")
event1.create_event()
event1.add_attendee("Alice Johnson")
event1.add_attendee("Bob Smith")
event1.remove_attendee("Alice Johnson")
print(f"Total attendees: {event1.get_total_attendees()}")


private_event = PrivateEvent("VIP Dinner", "2024-09-11", "8:00 PM", "Private Residence")
private_event.create_event()
private_event.check_invitation_status()
private_event.add_attendee("John Doe")

public_event = PublicEvent("Music Concert", "2024-09-12", "7:00 PM", "Stadium", max_capacity=3)
public_event.create_event()
public_event.add_attendee("Eve Davis")
public_event.add_attendee("Chris Martin")
public_event.add_attendee("David Lee")
public_event.add_attendee("Elon Musk")  
public_event.check_capacity()
```

```python
# 10.Write a code for Airline Reservation System using OOPS:
# Create a Flight class that has attributes such as flight number, departure and arrival airports, departure and 
# arrival times, and available seats (private).
# Implement methods to book a seat, cancel a reservation, and get the remaining available seats.
# Use Encapsulation to hide the flight's unique identification number.
# Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their 
# own specific attributes and methods.

class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_available_seat):
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seat = [0] * total_available_seat
        self.__unique_Fid = hash(flight_number)

    def check_available_seat(self):
        available_seat = self.__available_seat.count(0)
        if available_seat > 0:
            print("Seat is available.")
        else:
            print("Sorry! All seats are already occupied.")

    def booking_seat(self):
        available_seat = self.__available_seat.count(0)
        if available_seat > 0:
            index_seat = self.__available_seat.index(0)
            self.__available_seat[index_seat] = 1
            print("Your seat is booked successfully.")
            print(f"Your flight number is {self.flight_number} and your confirmed seat is {index_seat}. Departure airport is {self.departure_airport}.")
        else:
            print("Sorry! Seat is not available.")

    def cancel_reservation(self, seat_no):
        if 0 <= seat_no < len(self.__available_seat) and self.__available_seat[seat_no] == 1:
            self.__available_seat[seat_no] = 0
            print("Your seat is cancelled.")
            print(f"Your flight number is {self.flight_number} and your cancelled seat was {seat_no}.")
        else:
            print("Invalid seat number or seat is already available.")

    def get_available_seat_no(self):
        available_seats = [i for i in range(len(self.__available_seat)) if self.__available_seat[i] == 0]
        print("Available seat numbers are:")
        print(available_seats)

    def to_get_flight_unique_id(self):
        print(f"This flight has a unique identification number: {self.__unique_Fid}")

    def printing_flight_detail(self):
        print(f"Flight ID: {self.__unique_Fid}, Flight number: {self.flight_number}")
        print(f"Departure airport: {self.departure_airport}, Departure time: {self.departure_time}")
        print(f"Arrival airport: {self.arrival_airport}, Arrival time: {self.arrival_time}")
        available_seat = self.__available_seat.count(0)
        print(f"Available seats: {available_seat}")


class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_available_seat, domestic_fee):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_available_seat)
        self.domestic_fee = domestic_fee

    def print_flight_fee(self):
        print(f"Domestic flight fee: {self.domestic_fee}")


class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_available_seat, international_fee, visa_required):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_available_seat)
        self.international_fee = international_fee
        self.visa_required = visa_required

    def print_flight_fee(self):
        print(f"International flight fee: {self.international_fee}")

    def print_visa_requirement(self):
        print(f"For international travel, visa requirement: {self.visa_required}")

flight1 = Flight("101","delhi","patna","10.00 AM","20:00 AM",60)
flight1.check_available_seat()
flight1.booking_seat()
flight1.cancel_reservation(0)
flight1.get_available_seat_no()
flight1.to_get_flight_unique_id()
flight1.printing_flight_detail()

domestic_flight = DomesticFlight('DL123', 'JFK', 'LAX', '10:00Am', '13:00 pm', 100, 150)
international_flight = InternationalFlight('AI456', 'JFK', 'LHR', '20:00pm', '08:00am', 200, 1000, True)

domestic_flight.printing_flight_detail()
domestic_flight.print_flight_fee()

            
```

```python
# . Calculate coefficient of correlation between the marks obtained by 10 students in Accountancy and 
# statistics: Use Karl Pearson’s Coefficient of Correlation Method to find it
import numpy as np


X = np.array([45,70,65,30,90,40,50,75,85,60])
Y = np.array([35,90,70,40,95,40,60,80,80,50])


mean_X = np.mean(X)
mean_Y = np.mean(Y)


dX = X - mean_X
dY = Y - mean_Y


dX_squared = dX ** 2
dY_squared = dY ** 2


dX_times_dY = dX * dY


sum_dX_squared = np.sum(dX_squared)
sum_dY_squared = np.sum(dY_squared)
sum_dX_times_dY = np.sum(dX_times_dY)


r = sum_dX_times_dY / np.sqrt(sum_dX_squared * sum_dY_squared)

print(f"Coefficient of Correlation (r): {r:.3f}")
```

```python
# Make a fully functional web application using flask, Mangodb. Signup,Signin page.And after successfully 
# login .Say hello Geeks message at webpage.

from flask import Flask, render_template, redirect, url_for, request, flash
from flask_pymongo import PyMongo
from flask_wtf import FlaskForm
from wtforms import StringField, PasswordField, SubmitField
from wtforms.validators import DataRequired, Length
from werkzeug.security import generate_password_hash, check_password_hash

app = Flask(__name__)
app.config['SECRET_KEY'] = 'your_secret_key'  # Change this to a strong secret key
app.config['MONGO_URI'] = 'mongodb://localhost:27017/mydatabase'

mongo = PyMongo(app)

class SignupForm(FlaskForm):
    username = StringField('Username', validators=[DataRequired(), Length(min=4, max=25)])
    password = PasswordField('Password', validators=[DataRequired(), Length(min=6, max=35)])
    submit = SubmitField('Sign Up')

class SigninForm(FlaskForm):
    username = StringField('Username', validators=[DataRequired()])
    password = PasswordField('Password', validators=[DataRequired()])
    submit = SubmitField('Sign In')

@app.route('/signup', methods=['GET', 'POST'])
def signup():
    form = SignupForm()
    if form.validate_on_submit():
        username = form.username.data
        password = generate_password_hash(form.password.data)
        
        existing_user = mongo.db.users.find_one({'username': username})
        if existing_user:
            flash('Username already exists.')
            return redirect(url_for('signup'))

        mongo.db.users.insert_one({'username': username, 'password': password})
        flash('Signup successful! You can now sign in.')
        return redirect(url_for('signin'))
    
    return render_template('signup.html', form=form)

@app.route('/signin', methods=['GET', 'POST'])
def signin():
    form = SigninForm()
    if form.validate_on_submit():
        username = form.username.data
        password = form.password.data
        
        user = mongo.db.users.find_one({'username': username})
        if user and check_password_hash(user['password'], password):
            return redirect(url_for('hello'))
        
        flash('Invalid username or password.')
    
    return render_template('signin.html', form=form)

@app.route('/hello')
def hello():
    return 'Hello Geeks'

if __name__ == '__main__':
    app.run(debug=True)
```

```python
 # What is the difference between supervised and unsupervised learning ?
# Supervised Learning :->The ML style where Dependent Variable(Target) are present.
#                        In this learning Historical Data are present.
#                        Historical Learning act as supervisor (in this type of learning).
# UnSupervised Learning :-> The ML style where Dependent Variable(Target) are not Present.
#                           In this learning Historical Data are not present.
#                           So, Outcome Variable (y) are not present in this type of learning.
```

```python
 # Exlplain the bias-variance tradeoff.
# Bias-variance Tradeoff :->
# 1. It is a concept in ML describes the tradeoff between two types of errors(bias and variance) when creating a predictive model.
# 2. It has X-axis(Complexity of model) and Y-axis(Bias,Variance,Total_error).
# 3.It has Four reason.
#                    a. High Bias and Low Variance  -> Underfitting
#                    b. Low Bias and Low Variance -> Best Generalised Model
#                    c. High Variance and low Bias -> Overfitting
#                    d. High Variance and High Bias -> Does not make sense
# 4. TradeOff Point :-> optimal model Complexity where bias and Variance are balanced and also Total Error is minimum.
```

```python
 # What are precision and recall? How are they difference from accuracy?

# Precision :-> In all predicted ones how many ones are correctly predicted.
#               Here FP (False Positive)  are important.
# Recall :-> In all actual ones how many are correctly predicted.
#             Here FN(False Negative) is important.
# Accuracy :-> out of all Predicted Data How many( either zeros and ones) are correctly predicted.
#              Here FN and FP are important.
```

```python
# What is overfitting and how can it be prevented?

# Overfitting :-> In this fitting Bias is low but Variance is High. 
#                 On Training Data model having High accuracy but On Testing Data model having low accuracy.
#                 It Means That model is memorising the Data.
# In Regression,Classification, For performing well on trained Data for also performing on test Data well we have Regularisation Penalty,cross Validation.
# In Decision Tree, for performing well on trained Data for also performing on test Data Well we have Preprunning,PostPrunning,Ensambles and cross validation.
# In SVM,we have Regularisation Parameter(c).selection of kernel,cross validation.
                                      
```

```python
# Explain the concept o# ensemble learning.
# Ensembles :-> Combine multiple model prediction more stable and accurate as compared to individual model.
#               In this, Not necessary only one type of algorithm.
#               We have two type of technique to make ensembles model
#               1. Parallel Technique(Bagging)
#               2. Sequencial Technique(Boosting)
# Parallel Technique :-> in this we have same data but on this same data we make model using different algorithm.
#                        All of the model here are build parallely and independent of each other.
# Sequencial Technique :-> In this we have same data on this same data we make model using different algorithm.
#                          But It is specially used for decision Tree.
#                          All the model are build sequentially and dependent on each other.
#                          In DT1 has some misclassification then in DT2 has focus on misclassification of DT1.
#                          We we reach DT3, In DT2 has some misclassification then DT3 has focus on misclassification of DT2.
#                          We can say that it is learning from mistake
```

```python
# What is gradient descent and how does it work ?

# Gradient Descent :-> for model best accuracy we have to select the best line .To selecting the Best Fit line we use gradient descent.
#                      In simple regression, we want that combination of (m,c) so that overall error is least. 
#                      we used Gradiant Descent concept to finding optimal (m,c).
# working step :-> 
# 1. Start with a intial guess of model parameter(m,c).Guess is randomly. at with this parameter,we calculate (MSE).
# 2. calculate the gradient of cost function with respect to each parameter.
# 3. this process which calculating gradient repeat untill convergence.
#    θ:=θ−α⋅∇J(θ) (theta is can be slop and intercept. alpha is learning Rate it should be minimum.

# when slope is negative then:
#                              new(M) is greater than old(M)
#                              stop when new(M) should be zero.
# when slope is positive then:
#                            new(M) is lesser than old(M).
#                            stop when new(M) should be zero.
                               
```

```python
# Describe the difference between batch gradient descent and stochastic gradient descent.
# Batch Gradient Descent :-> 
#                            Uses the entire training dataset to compute the gradient of the cost function and update the model parameters in each iteration.
# Stochastic Gradient Descent :-> 
#                            Uses a single data point (or a small batch of data points) to compute the gradient and update the model parameters in each iteration.                      
```

```python
# # Explain the basic concept of a Support Vector Machine (SVM)
# Support Vector Machine (SVM) :-> Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. 
#                                  The basic idea behind SVM is to find the optimal boundary (or hyperplane) that best separates the classes in the feature space.
# Support Vector Classifier :-> It is say that we make to draw a line boundary such  that it is equidistance from two nearest points of two different class.
# Support Vector Regressor :-> to find a line (or hyperplane in higher dimensions) that best fits the data, while ensuring that as many data points as possible are within a certain distance from this line.
                                
```

```python
 # How does the kernel trick work in SVM ?
#  Kernel Trick :-> Kernal trick is fundamental concept in SVM that allow them to efficiently handle non linear classification task without explicitly mapping the data into a higher dimentional space.
# Working :->
#            If the data points are mixed in such a way that no straight line can separate the circles from the crosses, a simple SVM won’t work well.
#            The kernel trick "lifts" or "transforms" these data points into a higher-dimensional space where it might be easier to separate them.
#            he kernel function is the mathematical tool that allows this transformation without explicitly moving every data point to the higher dimension.
#            Once the data is "lifted" to this higher-dimensional space using the kernel function, the SVM can find a hyperplane that separates the classes.
```

```python
# What are the different types of kernels used in SVM and when would you use each ?
# there  are two type of kernal which is used in SVM:-
# 1.  Polynomial Kernel :-> The polynomial kernel computes the dot product the dot product between two vectors in a higher dimensional space using polynomial function.
#                           Useful when the decision boundary is non-linear and you suspect that a polynomial relationship might exist between features.

# 2. Gaussian Kernel :->   the gaussian kernel also known as Radial Basic Func transform the data into a infinite dimensional space.
#                          When the data is not linearly separable, and you need a highly flexible decision boundary
```

```python
#  What is the hyperplane in SVM and how is it determined ?

# hyperplane in SVM :-> 
#                      it is plane that seperate the data points of different classes.In 2D it is a line, But in higher dimensional it is hyperplane.
# How to determined :->
#                     The SVM algorithm finding Hyperplane by maximize the margin. margin is the distance between the two plane which is passes through the 
#                     support vectors of each classes.
#                     This SVM algorithm ensure that data are clearly classified in  particularly in higher dimensional space.
```

```python
# What are the pros and cons o# using a Support Vector Machine (SVM)?
# Pros of SVM :-> 
# 1. It is better version of Classification.
# 2. In this it make a decision boundary such that it is equidistance from two or more than two data points of opposite class
# 3. SVM works well even when the number of features is greater than the number of samples.
# 4. By using kernel functions, SVM can handle non-linear relationships in the data.
# cons of SVM :->
# 1. If classes are heavily overlapping, SVM may struggle to find a good decision boundary.
# 2. SVM is sensitive to noise (outliers), which can affect the position of the hyperplane and reduce accuracy.
# 3. SVM can be slow and memory-intensive for large datasets
```

```python
 # Describe the working principle o# a decision tree.
# Process :-> 
# 1. select the best variable for split (select through Information gain ).
# 2. Apply the split
# 3. Repeat the process for the subset obtained
# 4. Continue the process one and Two until every node is pure/stopping crieteria is reached.
# 5. Majority value in the leaf node will be prediction.
```

```python
# Describe the process of constructing a decision tree.
# Begin with all the data points in your dataset.
# This selection is based on a criterion like Gini impurity or information gain (using entropy), which measure how well the feature splits the data.
# Divide the dataset into subsets based on the chosen feature's values.
# For each subset, repeat the process of selecting the best feature to split on and creating further branches.
# Once no further splits can be made, the nodes become leaf nodes, which represent the final classification or decision.
```

```python
# What is information gain and how is it used in decision trees.
# Information Gain :-> It is the diffrence between impurity(Pre split) and impurity(Post split).

# How it is used in Decision Tree:->
#                                   which feature having higher Information Gain that will be selecting for spliting.Which feature having less Information gain
#                                   that will be selecting for spliting.
```

```python
 # Explain Gini impurity and its role in decision trees
# Gini Impurity :-> Gini impurity quantifies the likelihood of incorrectly classifying a randomly chosen element from the dataset if it were randomly labeled according to the distribution of labels in the dataset.
# Role in DT:->
#             Gini impurity helps in measuring how good a split is.
#             The process of calculating Gini impurity and selecting the best split is repeated for each node until the stopping criteria are met
#             During the construction of a decision tree, the algorithm evaluates each possible feature and threshold for splitting the data
```

```python
 # What are the advantages and disadvantages o# decision trees?
# Advantages:->
# 1. Decision trees are intuitive and easy to interpret, even for people without a technical background.
# 2. Decision trees can work with both types of data, making them versatile for different kinds of datasets.
# 3. decision trees don't require normalization or standardization of data.
# 4. Decision trees can capture complex, non-linear relationships between features and the target variable.
# Disadvantages :->
# 1.Decision trees can easily become too complex and overfit the training data, especially if they are not pruned or if the tree grows too deep.
# 2.Decision trees use a greedy approach, making locally optimal decisions at each step.
# 3.If one class is more frequent than others, the decision tree might become biased towards that class, making it less accurate.
```

```python
# How do random forests improve upon decision trees?
# Random forests improve upon decision trees by combining multiple decision trees to make predictions
# Each tree in the forest is trained on different subsets of the data and features, which reduces overfitting and increases accuracy.
# The final prediction is made by averaging the results or using majority voting from all trees.
```

```python
# How does a random forest algorithm work?

# Randomly select subsets of the data (with replacement) to train each decision tree.
# For each split in a tree, randomly select a subset of features to consider. This helps trees be diverse and reduces correlation among them.
# Train each decision tree on its respective data subset and feature subset. Each tree is grown to its maximum depth or until certain stopping criteria are met.
# For classification, take a majority vote from all the trees to determine the final class. 
```

```python
# What is bootstrapping in the context of random forests?
# bootstrapping is a method used to create multiple training datasets by randomly sampling from the original data with replacement.
# This means each subset may have repeated data points and some points may be left out. 
#  Each decision tree in the forest is trained on one of these bootstrapped datasets, helping to build diverse trees and improve the overall accuracy of the model.
```

```python
# Explain the concept of feature importance in random forests.
# Feature importance in random forests measures how much each feature (variable) contributes to the predictions made by the model.
# This is done by evaluating how much the model's accuracy decreases when the values of a feature are shuffled or removed.
# Features that lead to a significant drop in accuracy are considered more important.
```

```python
# What are the key hyperparameters of a random forest and how do they affect the model
# More trees generally improve model performance and stability but increase computation time.
# Determines the number of features to consider when looking for the best split. 
```

```python
# Describe the logistic regression model and its assumptions.
# Logistic regression is a statistical method used for binary classification tasks, where the goal is to predict the probability of a binary outcome (e.g., yes/no, 1/0). 
# Assump[tions :->
#     Assumes that the dependent variable is binary. Logistic regression is specifically designed for predicting binary outcomes.
#     Assumes that the independent variables are not highly correlated with each other. Multicollinearity can make it difficult to estimate the coefficients accurately.
#     Assumes that the observations (data points) are independent of each other.
```

```python
 # How does logistic regression handle binary classification problems?
# Logistic regression handles binary classification problems by estimating the probability that a given input belongs to one of the two classes. 
```

```python
# What is the sigmoid #unction and how is it used in logistic regression?
# he sigmoid function in logistic regression transforms the linear combination of input features into a probability value between 0 and 1.
# This probability represents the likelihood of an input belonging to the positive class. 
# By using a threshold (usually 0.5), the model can then classify the input into one of the two classes, effectively making binary predictions based on the calculated probability.
```

```python
#  Explain the concept of the cost function in logistic regression.

# In logistic regression, the cost function measures the error between predicted probabilities and actual class labels.
# It is typically the binary cross-entropy loss, which quantifies how well the model's predictions match the true labels. 
# The goal is to minimize this cost to improve the model's accuracy.
```

```python
# How can logistic regression be extended to handle multiclass classification
# Logistic regression can be extended to handle multiclass classification using the One-vs-Rest (OvR) approach,
# which trains separate binary models for each class, or by using Softmax Regression (Multinomial Logistic Regression),
# which generalizes the logistic function to output probabilities for all classes simultaneously.
# In both methods, the class with the highest predicted probability is chosen as the final output.
```

```python
# What is the difference between L1 and L2 regularization in logistic regression?
# L1 and L2 regularization in logistic regression both help prevent overfitting by adding a penalty to the model's coefficients. 
# L1 regularization (Lasso) adds the absolute value of the coefficients as a penalty, which can reduce some coefficients to zero, 
# effectively performing feature selection. L2 regularization (Ridge) adds the square of the coefficients as a penalty, 
# which reduces the size of all coefficients but usually keeps them non-zero, retaining all features.
```

```python
 # Explain the concept of boosting in the context of ensemble learning.
               # Boosting is an ensemble learning technique that aims to improve the performance of weak learners (models that perform slightly better than random guessing) by combining them into a strong, more accurate model.
               # The process works by sequentially training models.
               # Build a first model on training dataset and then build a second model to rectify the errors present in thie first model.
               # this procedure is continued untill and unless the errors are minimized,data is correctly predicted.
```

```python
# 1 How does XGBoost handle missing values ?
#                      XGBoost handles missing data by automatically learning the best path to take for missing values during tree construction, 
#                      so there's no need for imputation or special preprocessing.
```

```python
# What are the key hyperparameters in XGBoost and how do they affect model performance?
# In XGBoost, hyperparameter is lemda.
# Lemda is  give impact in calculating Similarity weight.
# If the lemda is increases then similarity weight decreased.
# If the lemda is decreases then similarity weight increased.
# whichever feature having highest similarity gain the split will happen on that feature
```

```python
# Describe the process of gradient boosting in XGBoost.
# Gradient Boosting in XGBoost :->
#                                 Gradient boosting in XGBoost is a process where models are sequentially trained, 
#                                 with each new model focusing on correcting the errors of the previous ones.
#                                 It works by minimizing a loss function using gradient descent, 
#                                 where the new model fits the residuals (errors) of the combined predictions from earlier models.
```

```python
# What are the advantages and disadvantages of using XGBoost?
# Advantage :->
#             1. Optimized for fast computation, making it efficient for large datasets.
#             2. Automatically manages missing values effectively.
#             3. Supports various objective functions and evaluation metrics.
# Disadvantages :->
#                  1. Can be difficult to tune and interpret.
#                  2. Risk of overfitting, especially on small or noisy datasets, if not properly tuned.
#                  3. Requires more memory and computational power compared to simpler models.
```
