```python
def size_object_to_intger(x):
    if(x[-1]=='M'):
        x=float(x.replace("M",""))
        x=x*1024
        return x
    elif(x[-1]=='k'):
        x=float(x.replace('k',''))
        return x
    else:
        return np.NaN
         
```

```python
df['Size']=df['Size'].apply(size_object_to_intger)
```
